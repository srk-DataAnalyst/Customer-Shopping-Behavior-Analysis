```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# Create a column age_group
labels = ['Young Adult','Adult','Middle-Aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# Create column purchase_frequency_days
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Annually' : 365,
    'Bi-Weekly' : 14,
    'Every 3 Months' : 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

Push CSV to MSSQL

```python
from sqlalchemy import create_engine
import pandas as pd

SERVER_NAME = r'SRK_ROG\SRK_ROG_SQL'
DATABASE_NAME = 'customer_shopping_behavior'
TABLE_NAME = 'customer'

# Create SQL Server engine (Windows Authentication)
engine = create_engine(
    f"mssql+pyodbc://@{SERVER_NAME}/{DATABASE_NAME}"
    "?driver=ODBC+Driver+17+for+SQL+Server&trusted_connection=yes"
)

# Push cleaned dataframe to SQL Server
df.to_sql(
    TABLE_NAME,
    engine,
    if_exists="replace",
    index=False
)

# Verify insert
pd.read_sql(f"SELECT TOP 5 * FROM {TABLE_NAME};", engine)
```
